```python
from PIL import Image, ImageDraw

# 이미지 객체로 생성
im = Image.open('/content/take_a_delicious_coffee_test.png')

# RGB 모드로 변경
rgb_im = im.convert('RGB')

# 지정한 좌표(1, 400)의 색상을 r,g,b 변수에 넣음
r, g, b = rgb_im.getpixel((1, 400))

print(r)
print(g)
print(b)
```

```python
# 누끼컷으로 변경할 이미지
img = Image.open('./take_a_delicious_coffee_test.png')
img = img.convert("RGBA")
datas = img.getdata()
newData = []
for item in datas:
    if item[0] == 255 and item[1] == 255 and item[2] == 255:
        newData.append((255, 255, 255, 0))
    else:
        newData.append(item)
img.putdata(newData)

# png 확장자로 저장
img.save("./TransparentImage.png", "PNG")
```

```python
template = Image.open("./salad_onthe_browntable.png")
rgb_im = template.convert('RGB')

proText = Image.open("./TransparentImage.png")

# template 이미지 객체 0,0 위치에 proText 이미지를 붙여줍니다.
rgb_im.paste(proText, (0, 0))

# 붙인 이미지를 새로운 이름으로 저장합니다.
rgb_im.save("./pasted_proText.jpg", "JPEG")
```
